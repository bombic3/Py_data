```python
# 부동 소수점 사용
# 평수 구하는 방법
SQMETER_PER_P = 3.3

# area = float(input("면적(제곱미터) : "))
area = eval(input("면적(제곱미터) : ")) # eval => 데이터에 따라 실수 또는 정수로 변환
py = area / SQMETER_PER_P
print(py, "평")
```

```python
# 형식화된 출력
SQMETER_PER_P = 3.3

area = eval(input("면적(제곱미터) : ")) # eval => 데이터에 따라 실수 또는 정수로 변환
py = area / SQMETER_PER_P
print("%.2f" % py, "평")
```

```python
# BMI 계산하기
# BMI < 18.50 저체중
# BMI 18.50 ~ 24.99 표준
# BMI 24.99 ~ 29.99 과체중
# BMI > 30 비만

weight = float(input("몸무게 kg 단위를 입력하세요 : "))
height = float(input("키를 cm 단위로 입력하세요 : "))

bmi = (weight/(height**2))*10000
print("당신의 BMI = ", bmi)
```
