```python
# 원리금 합계 구하기
a = 1000   # 원금
r = 0.05   # 이자율
n = 10   # 기간
result = a*(1+r)**n    # 원리금 합계(** -> 승)

print("원리금 합계 = ", result)
```

```python
# 소수점 자리(round(이름,자릿수))
price = 12345
tax = price * 0.075
tax = round(tax ,2) # 소수점 두자리까지
print(tax)
```
